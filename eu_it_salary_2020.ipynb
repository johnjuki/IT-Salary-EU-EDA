## Salary

```python
sns.set_style('whitegrid')
```

```python
sns.boxplot(df['Yearly brutto salary (without bonus and stocks) in EUR']);
```

There are outliers from this box plot, so we have to handle them

```python
# Removing outlier using IQR

base_salary = 'Yearly brutto salary (without bonus and stocks) in EUR'

p25 = df[base_salary].quantile(0.25) 
p75 = df[base_salary].quantile(0.75) 
iqr = p75 - p25 

upper_limit = p75 + 1.5 * iqr
lower_limit = p25 - 1.5 * iqr 

# New DataFrame after removing the outlier 
new_df = df[(df[base_salary] > lower_limit) & (df[base_salary] < upper_limit)]
```

```python
# Distribution after removing outlier from the dataset

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,6))

sns.histplot(new_df[base_salary], ax=ax1);
sns.boxplot(new_df[base_salary],ax=ax2);

plt.tight_layout()
```

## Age


```python
# Count plot
plt.figure(figsize=(15,8))
sns.countplot(data=new_df, x= 'Age')

plt.xticks(rotation=90)
plt.ylabel("Number of respondents");
```

## Gender

```python
new_df['Gender'].unique()
```

```python
# Replacing NaN value with string 'Not Specified'
new_df['Gender'] = new_df['Gender'].fillna(value='Not Specified')
new_df['Gender'].unique()
```

```python
plt.figure(figsize=(15,8))

new_df['Gender'].value_counts().plot.pie(autopct="%1.2f%%",fontsize=12,startangle=90,
                                         explode=[0.05] * 3,pctdistance=1.1,
                                          labeldistance=1.3,textprops={'fontsize': 15})

plt.ylabel("");
```

## Position


```python
# Removing the space at the end of the position column

new_df.rename(columns={'Position ': 'Position'}, inplace = True)
new_df['Position'].value_counts()
```

### Salary and position


```python
plt.figure(figsize=(15,8))

sns.countplot(data=new_df, x= 'Position',order= new_df['Position'].value_counts().iloc[:10].index);
```

```python
position_earning = new_df[base_salary].groupby(new_df['Position']).sum()
position_earning.sort_values(ascending=False,inplace=True)

position_earning = position_earning.to_frame()

plt.figure(figsize=(12, 8))

sns.barplot(data = position_earning, y = position_earning.index[:10], x= position_earning[base_salary].iloc[:10])

plt.ticklabel_format(style='plain', axis='x');
```
